```python
# configuration
import matplotlib.pyplot as plt

free = ["ended_results_", "ended_results_goSide_", "ended_results_goback_"]
grid = ["ended_results_grid_", "ended_results_grid_goSide_", "ended_results_grid_goback_"]
size = ["10", "20", "30", "40", "50", "60", "70"]
size_int = range(10, 80, 10)
robots = ["2", "5", "7", "10"]
points = ["3", "8"]
options_free = ["100", "101", "201", "110", "111", "211"]
options_grid = ["110", "111", "211"]
```

```python
# time split

for name in free:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_time/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'w') as f:
                f.write("")

for name in grid:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_time/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'w') as f:
                f.write("")

for name in free:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_time/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'a') as f:
                f.write(data[0] + " " + data[8] + "\n")

for name in grid:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_time/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'a') as f:
                f.write(data[0] + " " + data[8] + "\n")
    
```

```python
# percent split

for name in free:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_percent/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'w') as f:
                f.write("")

for name in grid:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_percent/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'w') as f:
                f.write("")

for name in free:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_percent/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'a') as f:
                f.write(data[0] + " " + str(int(data[9])/int(data[7])*100) + " " + str(int(data[5])/(int(data[0])*int(data[0])*int(data[2])/100)*int(data[7])) + "\n")

for name in grid:
    file_name = "ended_results/" + str(name) + ".txt"
    with open(file_name, 'r') as f:
        for line in f:
            data = line.split(" ")
            with open("finish_percent/" + str(name) + data[2] + "_" + data[3] + "_" + data[4] + ".txt", 'a') as f:
                f.write(data[0] + " " + str(int(data[9])/int(data[7])*100) + " " + str(int(data[5])/(int(data[0])*int(data[0])*int(data[2])/100)*int(data[7])) + "\n")
```

```python
time_results = {}

for name in free:
    time_results[name] = {}
    for robot in robots:
        time_results[name][robot] = {}
        for point in points:
            time_results[name][robot][point] = {}
            for option in options_free:
                time_results[name][robot][point][option] = []
                file_name = "finish_time/" + str(name) + str(robot) + "_" + str(point) + "_" + str(option) + ".txt"
                with open(file_name) as f:
                    for line in f:
                        data = line.split(" ")
                        time = data[0]
                        time_results[name][robot][point][option].append(time)


for name in grid:
    time_results[name] = {}
    for robot in robots:
        time_results[name][robot] = {}
        for point in points:
            time_results[name][robot][point] = {}
            for option in options_grid:
                time_results[name][robot][point][option] = []
                file_name = "finish_time/" + str(name) + str(robot) + "_" + str(point) + "_" + str(option) + ".txt"
                with open(file_name) as f:
                    for line in f:
                        data = line.split(" ")
                        time = data[0]
                        time_results[name][robot][point][option].append(time)
```

```python
percent_results = {}

for name in free:
    percent_results[name] = {}
    for robot in robots:
        percent_results[name][robot] = {}
        for point in points:
            percent_results[name][robot][point] = {}
            for option in options_free:
                percent_results[name][robot][point][option] = []
                file_name = "finish_percent/" + str(name) + "_" + str(robot) + "_" + str(point) + "_" + str(option) + ".txt"
                with open(file_name) as f:
                    for line in f:
                        data = line.split(" ")
                        time = data[0]
                        percent_results[name][robot][point][option].append(time)

for name in grid:
    percent_results[name] = {}
    for robot in robots:
        percent_results[name][robot] = {}
        for point in points:
            percent_results[name][robot][point] = {}
            for option in options_grid:
                percent_results[name][robot][point][option] = []
                file_name = "finish_percent/" + str(name) + "_" + str(robot) + "_" + str(point) + "_" + str(option) + ".txt"
                with open(file_name) as f:
                    for line in f:
                        data = line.split(" ")
                        time = data[0]
                        percent_results[name][robot][point][option].append(time)
```

```python
# graph time free
x = []
y = []

for name in free:
    for robot in robots:
        for point in points:
            for option in options_free:
                with open("finish_time/" + str(name) + "_" + str(robot) + "_" + str(point) + "_" + str(option) + ".txt") as f:
                    x = []
                    y = []
                    for line in f:
                        data = line.split(" ")
                        x.append(int(data[0]))
                        y.append(float(data[1]))
                    if int(point) == 3:
                        plt.plot(x, y, label=str(robot) + "% " + str(point) + " " + str(option))
                    else:
                        plt.plot(x, y, linestyle='--', label=str(robot) + "% " + str(point) + " " + str(option))
        plt.grid(linestyle='--')
        plt.title("Time of performing simulation - " + name)
        plt.xlabel("size [NxN]")
        plt.ylabel("time [ms]")
        plt.legend()
        plt.show()
```
